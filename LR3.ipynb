### Импортирт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
import yfinance as yf
```

### Загрузка временного ряда

```python
def load_data(tickers, start_date, end_date):
    data = yf.download(tickers,start_date,end_date)
    return data['Close']
```

```python
tsla=load_data('TSLA','2021-01-01','2023-12-31')
```

### Подготовка данных

Функция для разделения данных на настоящее и будущее

```python
def create_features(data, look_back=60):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back),0])
        y.append(data[i + look_back,0])
    return np.array(X), np.array(y)
```

Нормализуем данные

```python
scaler = MinMaxScaler()
tsla_scaled = scaler.fit_transform(tsla.values.reshape(-1, 1))
print(tsla_scaled.shape)
```

Разделение данных на тренировочную и тестовую выборки

```python
X, y = create_features(tsla_scaled)
print(X.shape,y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

### Модели машинного обучения

### Логистическая регрессия

Перебор гиперпараметров

```python
y_direction = (y_train > np.roll(y_train, 1))[1:].astype(int)
X_train_lr = X_train[1:]

lr_classifier = LogisticRegression(random_state=42)

lr_params={
    "tol":[0.00001,0.0001,0.001],
    "solver":['lbfgs','liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
lr_grid = GridSearchCV(lr_classifier, lr_params, cv=5, n_jobs=-1,verbose=3)
lr_grid.fit(X_train_lr, y_direction)
best_lr_params = lr_grid.best_params_
```

Лучшие гиперпараметры

```python
print(best_lr_params)
```

Тренировка модели с лучшими гиперпараметрами

```python
lr_model = LogisticRegression(**best_lr_params,random_state=42)
lr_model.fit(X_train_lr, y_direction)
```

### Машина опорных векторов (SVM)

Перебор гиперпараметров

```python
svm_classifier = SVR()

svm_params={
    "kernel":['linear', 'poly', 'rbf', 'sigmoid'],
    "degree":[1,3,5,7],
    "gamma": ['scale','auto'],
    "tol":[0.00001,0.0001,0.001]
}
svm_grid = GridSearchCV(svm_classifier, svm_params, cv=5, n_jobs=-1,verbose=3)
svm_grid.fit(X_train, y_train.ravel())
best_svm_params = svm_grid.best_params_
```

Лучшие гиперпараметры

```python
print(best_svm_params)
```

Тренировка модели с лучшими гиперпараметрами

```python
svm_model = SVR(**best_svm_params)
svm_model.fit(X_train, y_train.ravel())
```

### Random Forest

Перебор гиперпараметров

```python
rf_classifier = RandomForestRegressor()

rf_params={
    "n_estimators":[10,100,1000],
    "criterion":['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
    "max_depth":[1,10,100]
}
rf_grid = GridSearchCV(rf_classifier, rf_params, cv=5, n_jobs=-1,verbose=3)
rf_grid.fit(X_train, y_train.ravel())
best_rf_params = rf_grid.best_params_
```

Лучшие гиперпараметры

```python
print(best_rf_params)
```

Тренировка модели с лучшими гиперпараметрами

```python
rf_model = RandomForestRegressor(**best_rf_params, random_state=42)
rf_model.fit(X_train, y_train.ravel())
```

### XGBoost

Перебор гиперпараметров

```python
xgb_classifier = XGBRegressor()

xgb_params={
    "n_estimators":[10,100,1000],
    "learning_rate":[1,0.1,0.01,0.001],
    "max_depth":[1,10,100]
}
xgb_grid = GridSearchCV(xgb_classifier, xgb_params, cv=5, n_jobs=-1,verbose=3)
xgb_grid.fit(X_train, y_train.ravel())
best_xgb_params = xgb_grid.best_params_
```

Лучшие гиперпараметры

```python
print(best_xgb_params)
```

Тренировка модели с лучшими гиперпараметрами

```python
xgb_model = XGBRegressor(**best_xgb_params, random_state=42)
xgb_model.fit(X_train, y_train.ravel())
```

### Оценка моделей

Функция для оценки моделей

```python
def evaluate_model(model, X_test, y_test, model_name):
    if model_name == 'Logistic Regression':
        # Для логистической регрессии прогнозируем направление движения
        y_pred_direction = model.predict(X_test)
        y_pred = np.where(y_pred_direction == 1, y_test + 0.01, y_test - 0.01)
    else:
        y_pred = model.predict(X_test).reshape(-1, 1)
    
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    print(f'{model_name} - MAE: {mae:.4f}, RMSE: {rmse:.4f}')
    return y_pred
```

Сравнение результатов оценки разных моделей

```python
lr_pred = evaluate_model(lr_model, X_test, y_test, 'Logistic Regression')
svm_pred = evaluate_model(svm_model, X_test, y_test, 'SVM')
rf_pred = evaluate_model(rf_model, X_test, y_test, 'Random Forest')
xgb_pred = evaluate_model(xgb_model, X_test, y_test, 'XGBoost')
```

### Визуализация результатов

```python
plt.figure(figsize=(12, 6))
plt.plot(tsla.index[-len(y_test):], scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual')
plt.plot(tsla.index[-len(y_test):], scaler.inverse_transform(lr_pred.reshape(-1, 1)), label='Logistic Regression')
plt.plot(tsla.index[-len(y_test):], scaler.inverse_transform(svm_pred.reshape(-1, 1)), label='SVM')
plt.plot(tsla.index[-len(y_test):], scaler.inverse_transform(rf_pred.reshape(-1, 1)), label='Random Forest')
plt.plot(tsla.index[-len(y_test):], scaler.inverse_transform(xgb_pred.reshape(-1, 1)), label='XGBoost')
plt.title('Сравнение прогнозов моделей машинного обучения')
plt.xlabel('Дата')
plt.ylabel('Цена акции')
plt.legend()
plt.show()
```

Исходя из полученного графика, можно сказать что все модели сделали примерно похожие прогнозы, модель SVM больше всех имеет отличия в прогнозе

### Анализ важности признаков для Random Forest

```python
feature_importance = pd.DataFrame({'feature': range(X_train.shape[1]), 'importance': rf_model.feature_importances_})
feature_importance = feature_importance.sort_values('importance', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(feature_importance['feature'], feature_importance['importance'])
plt.title('Топ-10 важных признаков (Random Forest)')
plt.xlabel('Индекс признака')
plt.ylabel('Важность')
plt.show()
```

Видно, что в основном на прогноз влияют только значения за последние 3 периода перед предсказанием

```python
feature_importances = xgb_model.get_booster().get_score(importance_type='weight')
sorted_features = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)
importance_df = pd.DataFrame(sorted_features, columns=['Feature', 'Importance'])

print("Топ 10 признаков по важности:")
print(importance_df.head(10))
plt.figure(figsize=(10, 6))
importance_df.head(10).set_index('Feature').plot(kind='bar', legend=False)
plt.title("Топ 10 важных признаков (XGBoost)")
plt.xlabel("Индекс признака")
plt.ylabel("Важность")
plt.show()
```
